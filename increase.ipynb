```python
import pandas as pd
import numpy as np
data_root = pd.read_csv("koreanair_stock.csv")
```

```python
data = data_root[['대비','종가']].replace(',', '', regex=True).astype(int)
```

```python
dif = []

#1프로 이상 3프로 미만 = 1, 3프로 이상 5프로 미만 = 2, 5프로 이상 3
for i in range(0, len(data)-1):
    if (data.loc[i,"대비"] > 0) & (abs(data.loc[i,"대비"]) >= \
                                 (data.loc[i + 1, "종가"] / 100) * 1) & (abs(data.loc[i,"대비"]) \
                                                                       < (data.loc[i + 1, "종가"] / 100) * 3):
        dif.append(1)
    elif (data.loc[i,"대비"] > 0) & (abs(data.loc[i,"대비"]) >= \
                                   (data.loc[i + 1, "종가"] / 100) * 3) & (abs(data.loc[i,"대비"]) \
                                                                         < (data.loc[i + 1, "종가"] / 100) * 5):
        dif.append(2)
    elif (data.loc[i,"대비"] > 0) & (abs(data.loc[i,"대비"]) >= \
                                   (data.loc[i + 1, "종가"] / 100) * 5):
        dif.append(3)
    elif (data.loc[i,"대비"] < 0) & (abs(data.loc[i,"대비"]) >= \
                                   (data.loc[i + 1, "종가"] / 100) * 1) & (abs(data.loc[i,"대비"])\
                                                                         < (data.loc[i + 1, "종가"] / 100) * 3):
        dif.append(-1)
    elif (data.loc[i,"대비"] < 0) & (abs(data.loc[i,"대비"]) >= \
                                   (data.loc[i + 1, "종가"] / 100) * 3) & (abs(data.loc[i,"대비"])\
                                                                         < (data.loc[i + 1, "종가"] / 100) * 5):
        dif.append(-2)
    elif (data.loc[i,"대비"] < 0) & (abs(data.loc[i,"대비"]) >= \
                                   (data.loc[i + 1, "종가"] / 100) * 5):
        dif.append(-3)
    else :
        dif.append(0)        
```

```python
#마지막 날자 추가
dif.append(0) 
```
